```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
df['Reason']
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(df,x='Reason')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour']=df['timeStamp'].apply(lambda x: x.hour)
df['Month']=df['timeStamp'].apply(lambda x: x.month)
df['Day of Week']=df['timeStamp'].apply(lambda x: x.dayofweek)
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week']=df['Day of Week'].map(dmap)
df.head()
```

```python
sns.countplot(df, x='Day of Week', hue='Reason')
```

```python
sns.countplot(df, x='Month', hue='Reason')
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
byMonth.reset_index()
sns.lmplot(data=byMonth.reset_index(), x='Month', y='twp')
```

```python
df['Date']= df['timeStamp'].apply(lambda x: x.date())
```

```python
byDate =df.groupby('Date').count()
byDate['twp'].plot()
plt.tight_layout
```

```python
df[(df['Reason'] == 'Traffic')].groupby('Date').count()['twp'].plot()
```

```python
df[(df['Reason'] == 'Fire')].groupby('Date').count()['twp'].plot()
```

```python
df[(df['Reason'] == 'EMS')].groupby('Date').count()['twp'].plot()
```

```python
newdf=df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
newdf
```

```python
sns.heatmap(data=newdf, cmap='magma')
```

```python
sns.clustermap(data=newdf, cmap='magma')
```

```python
newdfmonth=df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
newdfmonth
```

```python
sns.heatmap(data=newdfmonth, cmap='magma')
```

```python
sns.clustermap(data=newdf, cmap='magma')
```
